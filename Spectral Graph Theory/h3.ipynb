**Exercise 1**: Generating the adjacency matrix A for graphs of particular type. Write a script that generates the adjacency matrix A for each of the following graphs and prints this matrix onto the screen:
1. K5: 5-clique, or a fully connected simple, undirected graph of 5 nodes
2. K5.3: two disconnected components, C1 and C2, where C1 is a 5-clique K5 and C2 is a 3-clique K3
3. K5.3e: Almost the same as K53 but there is a single edge connecting the two components
4. B2.3: Complete bi-partite graph with n1=2 nodes in the first part and n2=3 nodes in the second part
5. S5: A 5-vertex star (one central "hub" node that connects to all the other "spoke" nodes)
6. P5: A simple path of 5 vertices


```python


## 1.creating adjacency matrix for a 5-clique fully connected simple graph 
C1 = Graph.Full(5)
Graph_k5=C1.get_adjacency()
print("Adjacency matrix for 5- clique fully connected simple graph with undirected edges Graph_k5,",Graph_k5)

## 2.creating adjacency matrix for  two disconnected components, C1 and C2, where C1 is a 5-clique K5 and C2 is a 3-clique K3
C2 = Graph.Full(3)
K5_3 = C1+C2
Graph_K5_3=K5_3.get_adjacency()
print("Adjacency matrix for two disconnected components, C1 and C2",Graph_K5_3)


## 3.creating adjacency matrix for a K5.3e, adding the edge (4,5) connecting the two  connected components 
K5_3e = C1 + C2
K5_3e.add_edge(4,5)
Graph_K5_3e =K5_3e.get_adjacency()
print("Adjacency matrix for graph_K5.3e",Graph_K5_3e)

## 4.creating adjacency matrix  for bi-partite graph with n1=2 nodes in the first part and n2=3 nodes in the second part 
Bipart = Graph.Full_Bipartite(2,3)
Graph_bipartite=Bipart.get_adjacency()
print("Adjacency matrix for bipartite with n1=2 & n2=3 ",Graph_bipartite)

## 5.creating adjacency matrix  for 5-vertex star 
star_5 = Graph.Star(5)
Graph_star=star_5.get_adjacency()
print("Adjacency matrix for 5-vertex star graph",Graph_star)

## 6.creating adjacency matrix  for simple path of 5 vertices
path_graph = Graph()
path_graph.add_vertices(5)
path_graph.add_edges([(0,1),(1,2),(2,3),(3,4)])
Graph_path=path_graph.get_adjacency()
print("Adjacency matrix for a simple path of 5 vertices ",Graph_path)


```

**Exercise 2**: Generating the degree matrix D for a given adjacency matrix A. Write a script that generates a degree matrix for each of the adjacency matrices in Exercise 1. Note that the degree matrix is a diagonal matrix where all the positions except for the diagonal are zero's. The diagonal elements correspond to the degrees of the corresponding nodes, namely Dii = degree(v_i).

```python
# creating the function to return the degree of vertices of a graph.
def degree_matrix(graph):
    v= len(graph.vs)
    degree_matrix = np.zeros(shape=(v,v))
    # degree_matrix = [[0 for _ in range(v)] for _ in range(v)]
    # degree_matrix[i][i] = [graph.degree(i) for i in range(v)]
    for i in range(v):
        degree_matrix[i][i]= graph.degree(i)
    return degree_matrix

## 1. Degree matrix for a 5-clique fully connected simple graph 
degree_matK5 = degree_matrix(C1)
print("Degree matrix for 5-clique fully connected simple graph",degree_matK5)

## 2. Degreematrix for two disconnected components, C1 and C2, where C1 is a 5-clique K5 and C2 is a 3-clique K3
degree_matK5_3 = degree_matrix(K5_3)
print("Degree matrix for K5_3 graph",degree_matK5_3)

## 3. Degree matrix for a K5.3e, adding the edge (4,5) connecting the two  connected components 
degree_matK5_3e = degree_matrix(K5_3e)
print("Degree matrix  for a K5.3e graph",degree_matK5_3e)

## 4. Degree  matrix for a bi-partite graph with n1=2 nodes in the first part and n2=3 nodes in the second part 
degree_bipartite = degree_matrix(Bipart)
print("Degree matrix  for the above bipartite graph",degree_bipartite)

## 5. Degree matrix for a 5-vertex star graph 
degree_star5 = degree_matrix(star_5)
print("Degree matrix  for the 5 -vertex star",degree_star5)


## 6. Degree  matrix for a simple graph with 5 edges 
degree_path = degree_matrix(path_graph)
print("Degree matrix  for the simple graph with 5 vertices",degree_path)
```

**Exercise 4**: Generating the graph spectrum, or the multiset of the eigenvalues of the graph adjacency matrix A. Write a script that calculates the eigenvalues of the graph adjacency matrix for each of the matrices in Exercise 1. Plot the eigenvalues in the increasing order of their values

```python
def eigenvalue(graph):
  y = np.sort(np.linalg.eigvals(graph.data))
  print(y)
  plt.plot(y)
  plt.ylabel("Eigenvalue")
  plt.show()

#  Graph spectrum for K5 graph
print("eigen values for K5_graph")
eigenvalue(Graph_k5)
#  Graph spectrum for K5_3_graph 
print("eigen values for K5_3_graph")
eigenvalue(Graph_K5_3)
#  Graph spectrum for K5_3e_graph
print("eigen values for K5_3e_graph")
eigenvalue(Graph_K5_3e)
#  Graph spectrum for bipartite B2.3 graph
print("eigen values for bipartite B2.3 graph")
eigenvalue(Graph_bipartite)
#  Graph spectrum for 5-vertex star
print("eigen values for 5-vertex star")
eigenvalue(Graph_star)
#  Graph spectrum for path graph
print("eigen values for path graph")
eigenvalue(Graph_path)





```

**Exercise 5**: Generating the graph spectrum, or the multiset of the eigenvalues of the graph Laplacian. Write a script that calculates the eigenvalues of the graph Laplacian for each of the graphs in Exercise 1. Plot the eigenvalues in the increasing order of their values. 

```python
def eigenvalue_lap(lap_data):
  # y = np.sort(np.linalg.eigvals(lap_data))
  y = np.around(np.sort(np.linalg.eigvals(lap_data)), decimals = 3)
  print(y)
  plt.plot(y)
  plt.ylabel("Laplacian Eigenvalue")
  plt.show()

#  Graph spectrum for laplacian K5 graph
print("Laplacian Eigenvalues for K5_graph")
eigenvalue_lap(laplacian_K5)
#  Graph spectrum for laplacian K5_3_graph 
print("Laplacian Eigenvalues for K5_3_graph")
eigenvalue_lap(laplacian_K5_3)
#  Graph spectrum for laplacian K5_3e_graph 
print(" Laplacian Eigenvalues for K5_3e_graph")
eigenvalue_lap(laplacian_K5_3e)
#  Graph spectrum for laplacian  bipartite B2.3 graph
print("Laplacian Eigenvalues for bipartite B2.3 graph")
eigenvalue_lap(laplacian_bipart)
#  Graph spectrum for  laplacian 5-vertex star graph
print("Laplacian Eigenvalues for 5-vertex star")
eigenvalue_lap(laplacian_star5)
#  Graph spectrum for laplacian path graph 
print("Laplacian Eigenvalues for path graph")
eigenvalue_lap(laplacian_path)

```
